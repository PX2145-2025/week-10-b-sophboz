```python
def Spectrum1(x, Fs):
    """
    Compute one-sided FFT spectrum similar to 'Spectrum1' used in class:
    - x: time-domain samples (length N)
    - Fs: sampling frequency (Hz)
    Returns:
    - freqs: frequency axis for positive frequencies (0 .. Fs/2) length N//2
    - X_pos: complex FFT values for positive frequencies (not scaled)
    - A: one-sided amplitude spectrum scaled so that a pure sine of amplitude 1
         shows a peak of ~1 at the signal frequency (i.e. multiply by 2/N).
    """
    N = len(x)
    X = fft(x)
    Npos = N // 2
    X_pos = X[:Npos]
    freqs = np.linspace(0, Fs/2, Npos, endpoint=False)
    # One-sided amplitude scaling: multiply by 2/N to account for energy split
    A = (2.0 / N) * np.abs(X_pos)
    return freqs, X_pos, A

# Frequencies to test
f_tests = [0.20, 0.22]

#Vary N while keeping T fixed
T_fixed = 1.0
N_list = [64, 256, 1024]

# Make plots for part (a)
for f0 in f_tests:
    plt.figure(figsize=(10,6))
    for N in N_list:
        t = np.linspace(0, T_fixed, N, endpoint=False)
        Fs = 1.0 / (t[1] - t[0])
        x = np.sin(2*np.pi*f0*t)
        freqs, X_pos, A = Spectrum1(x, Fs)
        plt.plot(freqs, A, label=f'N={N}, Fs={int(Fs)} Hz')
    plt.xlim(0, 2.0)   # focus on low frequencies where the tone is
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('One-sided amplitude (scaled)')
    plt.title(f'Part (a): One-sided spectra for f={f0} Hz, varying N (T={T_fixed}s)')
    plt.legend()
    plt.grid(True)

# Also show real/imag for a representative N for each f (use N=256)
for f0 in f_tests:
    N = 256
    t = np.linspace(0, T_fixed, N, endpoint=False)
    Fs = 1.0 / (t[1] - t[0])
    x = np.sin(2*np.pi*f0*t)
    freqs, X_pos, A = Spectrum1(x, Fs)
    plt.figure(figsize=(10,4))
    plt.subplot(1,2,1)
    plt.plot(freqs, np.real(X_pos))
    plt.title(f'Real(X) for f={f0} Hz (N={N})')
    plt.xlabel('Frequency (Hz)')
    plt.grid(True)
    plt.subplot(1,2,2)
    plt.plot(freqs, np.imag(X_pos))
    plt.title(f'Imag(X) for f={f0} Hz (N={N})')
    plt.xlabel('Frequency (Hz)')
    plt.grid(True)

# Vary T while keeping sampling rate Fs fixed
Fs_fixed = 64.0
T_list = [1.0, 4.0, 16.0]

for f0 in f_tests:
    plt.figure(figsize=(10,6))
    for T in T_list:
        N = int(Fs_fixed * T)
        t = np.linspace(0, T, N, endpoint=False)
        x = np.sin(2*np.pi*f0*t)
        freqs, X_pos, A = Spectrum1(x, Fs_fixed)
        plt.plot(freqs, A, label=f'T={T}s, N={N}')
    # zoom to show the area around the tone
    plt.xlim(max(0, f0 - 0.5), f0 + 0.5)
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('One-sided amplitude (scaled)')
    plt.title(f'Part (b): One-sided spectra for f={f0} Hz, varying T (Fs={Fs_fixed} Hz)')
    plt.legend()
    plt.grid(True)

# show all figures
plt.show()
```

```python
N = 256
T = 1.0
t = np.linspace(0, T, N, endpoint=False)
Fs = 1/(t[1]-t[0])
freqs = np.linspace(0, Fs, N, endpoint=False)

f_list = [0.20, 0.22]
phi = np.pi/3

for f0 in f_list:
    x = np.cos(2*np.pi*f0*t + phi)
    X = fft(x)

    plt.figure(figsize=(10,4))
    plt.plot(freqs, np.real(X), label="Real")
    plt.plot(freqs, np.imag(X), label="Imag")
    plt.title(f"Real and Imag parts of X for f={f0} Hz, phase={phi:.2f} rad")
    plt.xlabel("Frequency (Hz)")
    plt.grid(True)
    plt.legend()

    plt.figure(figsize=(10,4))
    plt.plot(freqs, np.abs(X), label="|X|")
    plt.plot(freqs, np.angle(X), label="arg(X)")
    plt.title(f"Magnitude and Phase of X for f={f0} Hz, phase={phi:.2f} rad")
    plt.xlabel("Frequency (Hz)")
    plt.grid(True)
    plt.legend()

plt.show()
```

```python
def TestData(N, Fs):
    dt = 1.0 / Fs
    t = np.arange(N) * dt

    a1, f1 = 1.0, 12.3
    a2, f2 = 0.5, 7.5
    a3, f3 = 0.2, 30.0

    x = (a1 * np.cos(2*np.pi*f1*t) +
         a2 * np.cos(2*np.pi*f2*t) +
         a3 * np.cos(2*np.pi*f3*t))
    return x
```

```python
Fs = 200.0
N_values = [512, 1024, 2048]

for N in N_values:
    x = TestData(N, Fs)
    X = fft(x)
    X_pos = X[:N//2]
    freqs = np.linspace(0, Fs/2, N//2, endpoint=False)

    # One-sided amplitude spectrum (correct scaling)
    A = (2.0 / N) * np.abs(X_pos)

    plt.figure(figsize=(8,4))
    plt.plot(freqs, A)
    plt.title(f"FFT magnitude spectrum (N={N})")
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("Amplitude")
    plt.grid(True)
    plt.xlim(0, 50)
    plt.show()
```
